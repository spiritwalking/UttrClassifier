```python
import json
import codecs

dialog_list = json.loads(codecs.open("data/dialog_release.json", "r", "utf-8").read())
document_list = json.loads(codecs.open("data/document_url_release.json", "r", "utf-8").read())
```

```python
dialog_list
```

```python
data = []

for dialog in dialog_list:
    document_id = dialog['document_id']
    topic = document_list[document_id]['topic']
    if topic in ['娱乐', '游戏', '健康']:
        continue
    else:
        text = dialog['content']
        sample = {'text': text, 'topic': topic}
        data.append(sample)

data[0]
```

```python
from datasets import load_dataset

dataset = load_dataset('kd_conv')

dataset
```

```python
dataset['train'][0]
```

```python
en2zh = {'travel': '旅行', 'film': '电影', 'music': '音乐'}
for dialog in dataset['train']:
    en = dialog['domain']
    sample = {'text': dialog['messages']['message'], 'topic': en2zh[en]}
    data.append(sample)

for dialog in dataset['test']:
    en = dialog['domain']
    sample = {'text': dialog['messages']['message'], 'topic': en2zh[en]}
    data.append(sample)

for dialog in dataset['validation']:
    en = dialog['domain']
    sample = {'text': dialog['messages']['message'], 'topic': en2zh[en]}
    data.append(sample)

data[0]
```

```python
with open('data_ascii.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=4)
```

```python
import torch
from transformers import BertTokenizerFast, pipeline

model = torch.load('model.pth')
model.to('cpu')

tokenizer = BertTokenizerFast.from_pretrained('bert-base-chinese')

classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
uttr = "很多人都说学校食堂的菜不仅难吃、油水和肉少。"
classifier(uttr)
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline, BertConfig

config = BertConfig.from_pretrained('uer/roberta-base-finetuned-chinanews-chinese', num_labels=6)
model = AutoModelForSequenceClassification.from_pretrained('uer/roberta-base-finetuned-chinanews-chinese',
                                                           config=config)
tokenizer = AutoTokenizer.from_pretrained('uer/roberta-base-finetuned-chinanews-chinese')

text_classification = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
text_classification("北京上个月召开了两会")

model
```

```python
from transformers import DistilBertForSequenceClassification , AutoTokenizer, pipeline

model_name = "liam168/c4-zh-distilbert-base-uncased"
class_num = 6
ts_texts = ["女人做得越纯粹，皮肤和身材就越好", "我喜欢篮球"]
model = DistilBertForSequenceClassification.from_pretrained(model_name, num_labels=class_num, ignore_mismatched_sizes=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
classifier(ts_texts[0])
classifier(ts_texts[1])
```
