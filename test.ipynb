```python
import json
import codecs

dialog_list = json.loads(codecs.open("data/dialog_release.json", "r", "utf-8").read())
document_list = json.loads(codecs.open("data/document_url_release.json", "r", "utf-8").read())
```

```python
data = []

for dialog in dialog_list:
    for uttr in dialog['content'][2:-2]:
        document_id = dialog['document_id']
        data.append([uttr, document_list[document_id]['topic']])

for d in data:
    if d[1] == '教育':
        print(d)
```

```python
from transformers import BertTokenizerFast

# 实例化一个tokenizer对象
tokenizer = BertTokenizerFast.from_pretrained("bert-base-chinese")

# 编码一段文本
text = "西瓜和苹果哪个好吃？"
encoded_input = tokenizer(text, return_tensors='pt')

# 解码
decoded_input = tokenizer.decode(encoded_input['input_ids'][0])
```
